# Column of Interest

We throw away other columns of this dataset and just focus on "review" columns and the "label" column.

```python
selected_columns =  df.loc[:, ["review", "policies_violated"]] 
```

```python
selected_columns.head()
```

# Cleaning the Review Column

```python
import string
import re
nltk.download('stopwords')
nltk.download('wordnet')
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
corpus = []
for i in range(0, 1525):
  lema = WordNetLemmatizer()
  review = selected_columns['review'][i]
  review = ''.join(i for i in review if ord(i)<128)
  review = review.lower()
  remove_num=re.sub(r'\d+','',review)
  remove_punc = remove_num.translate(str.maketrans('','',string.punctuation))
  remove_white_space =  remove_punc.strip()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  review = ([lema.lemmatize(word) for word in remove_white_space.split() if word not in set()])
  review = ' '.join(review)
  corpus.append(review)
```

```python
print(corpus)
```

# Replacing the review columns by the processed & clean reviews

```python
selected_columns[["review"]] = pd.DataFrame(corpus)
selected_columns[:10]
```

#Many Missing Values with NA or NAN

```python
na_rows = []
multi_labels_rows = []
nan_rows =[]
for i in range(len(selected_columns)):
  if selected_columns.loc[i, "policies_violated"] == "na":
        na_rows.append(i)
  else:
    try:
      if len(selected_columns.loc[i, "policies_violated"].split("\r"))>1:
        multi_labels_rows.append(i)
    except:
      nan_rows.append(i)
bad_rows = list(set(na_rows + nan_rows))
print("There are ", len(bad_rows), "bad indices \n")
print(bad_rows)
```

```python
for item in np.array(selected_columns):
  try:
    if item[1]!= "na" and len(item[1].split("\r")) > 1:
      print(item, "\n")
  except:
    continue
```

#Ignoring the reviews with missing labels

```python
rest_rows= [i for i in range(len(selected_columns)) if i not in bad_rows]
cleaned_data = selected_columns.loc[rest_rows,:]
cleaned_data
```

# Bag of Words(BoW)

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer() 
X_train_counts = vectorizer.fit_transform(corpus)
X=pd.DataFrame(X_train_counts.toarray())
```

# Encoding the dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(y)
```

#Splitting of the dataset into test and train

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#Training the Naive Bayes model on the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
```

#Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

#Confusion matrix and Accuracy

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
accuracy_score(y_test, y_pred)
```
